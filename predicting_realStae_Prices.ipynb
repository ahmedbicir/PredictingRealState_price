```python
import pandas as pd
import numpy as np

import seaborn as sns


import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {'Hours': [1, 2, 4, 5, 6], 'Marks': [50, 60, 80, 90, 100]}
df = pd.DataFrame(data)
print(df)
```

```python
x =df[['Hours']].values
y =df['Marks'].values
model = LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print(f"slope(M):{slope}")
print(f"Intercept(b):{intercept}")
```

```python
print(f"the equation of the regression line of Y:{slope}X + {intercept}")
```

```python
df['Predicted marks'] = model.predict(x)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
label_encoder = LabelEncoder()

# Apply LabelEncoder to all string columns
for column in X_train.select_dtypes(include=['object']).columns:
    X_train[column] = label_encoder.fit_transform(X_train[column])

# Now all string columns are converted to numeric
```

```python
# One-Hot Encoding for all string columns
X_train = pd.get_dummies(X_train, drop_first=True)

# Now all string columns are converted to numeric (binary columns)
```

```python
# Convert columns with numeric-like strings to actual numeric
X_train = X_train.apply(pd.to_numeric, errors='coerce')

# The 'coerce' option will convert invalid parsing (non-numeric strings) to NaN, so make sure to handle NaN values afterward
X_train = X_train.fillna(0)  # or another imputation strategy
```

```python
# Check the data types of all columns
print(X_train.dtypes)
```

```python
from sklearn.linear_model import LinearRegression
lr =LinearRegression()
lr.fit(X_train,y_train)
```
